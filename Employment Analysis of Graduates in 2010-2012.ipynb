## Men and Women Distribution in Different Major Categories

```python
recent_grads["Major_category"].unique()
```

```python
Men_categories = []
Women_categories = []
for cat in recent_grads["Major_category"].unique():
    Men_categories.append(recent_grads.loc[recent_grads["Major_category"] == cat,'Men'].sum())
    Women_categories.append(recent_grads.loc[recent_grads["Major_category"] == cat,'Women'].sum())
new_frame = pd.DataFrame(recent_grads["Major_category"].unique(),columns =['Major_category']) 
```

```python
print(Men_categories)
print(Women_categories)
print(recent_grads["Major_category"].unique())
print(new_frame)
new_frame["Women"] = Women_categories
new_frame["Men"] = Men_categories
print(new_frame)
```

```python
new_frame.plot.bar(x='Major_category', y='Women')
```

```python
new_frame.plot.bar(x='Major_category', y='Men')
```

We see that while women are mainly employed in Business, Arts, and Education, Men are mostly employed in Business and Engineering
